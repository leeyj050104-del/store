### 01-2) 

높이 4인 이진 트리에서 가질 수 있는 **최대 리프 노드 수**는 포화 이진 트리일 때이며,

- 높이 `h`인 포화 이진 트리의 리프 노드 수 = `2^(h-1)`

따라서 높이 4이면

### 01-3) 

높이 4인 포화 이진 트리에서

- 전체 노드 수 = `2^4 - 1 = 15`
- 리프 노드 수 = `2^(4-1) = 8`
- 내부 노드 수 = `15 - 8 = 7`

따라서 차이는

- `8 - 7 = 1`

**답: 1개 차이난다.**

### 01-4) 

성립한다.

높이가 `h`인 포화 이진 트리에서

- 전체 노드 수 = `2^h - 1`
- 리프 노드 수 = `2^(h-1)`
- 내부 노드 수 = `(2^h - 1) - 2^(h-1) = 2^(h-1) - 1`

따라서

- 리프 노드 수 - 내부 노드 수  
= `2^(h-1) - (2^(h-1) - 1)`  
= `1`

즉, **모든 포화 이진 트리에서 리프 노드 수는 내부 노드 수보다 항상 1개 많다.**

**답: 성립한다.**

## 02 

### 02-1) 

완전 이진 트리에서 리프 노드의 수는

- `⌈n / 2⌉`


이유: 완전 이진 트리를 배열처럼 번호 붙이면  
`1`번부터 `⌊n/2⌋`번까지는 내부 노드이고,  
그 다음 노드들인 `⌊n/2⌋ + 1`번부터 `n`번까지가 리프 노드가 된다.

따라서 리프 노드 수는

- `n - ⌊n/2⌋ = ⌈n/2⌉`

**답: `⌈n/2⌉`개**

### 02-2) 
완전 이진 트리에서

- 내부 노드 수 = `⌊n/2⌋`
- 리프 노드 수 = `⌈n/2⌉`

따라서 관계는 다음과 같다.

- `n`이 짝수이면: 리프 노드 수 = 내부 노드 수
- `n`이 홀수이면: 리프 노드 수 = 내부 노드 수 + 1

즉, **리프 노드 수는 내부 노드 수와 같거나 1개 더 많다.**

## 03

주어진 트리:
```text
        5
      /   \
     3     8
    /     / \
   1     7   9
        /
       6
```

### 03-1) 

전위 순회는 **루트 → 왼쪽 서브트리 → 오른쪽 서브트리** 순서이다.

따라서 방문 순서는

- 5 → 3 → 1 → 8 → 7 → 6 → 9

**답: 5, 3, 1, 8, 7, 6, 9**

### 03-2) 

중위 순회는 **왼쪽 서브트리 → 루트 → 오른쪽 서브트리** 순서이다.

따라서 방문 순서는

- 1 → 3 → 5 → 6 → 7 → 8 → 9

**답: 1, 3, 5, 6, 7, 8, 9**

### 03-3) 

후위 순회는 **왼쪽 서브트리 → 오른쪽 서브트리 → 루트** 순서이다.

따라서 방문 순서는

- 1 → 3 → 6 → 7 → 9 → 8 → 5

**답: 1, 3, 6, 7, 9, 8, 5**

### 03-4) 

중위 순회는 항상

1. 왼쪽 서브트리를 먼저 방문하고
2. 그 다음 루트를 방문하고
3. 마지막으로 오른쪽 서브트리를 방문한다.

따라서

- 먼저 방문하는 왼쪽 서브트리의 값들은 루트보다 모두 작고,
- 나중에 방문하는 오른쪽 서브트리의 값들은 루트보다 모두 크다.

또한 왼쪽/오른쪽 서브트리도 각각 BST이므로, 같은 논리로 각 서브트리 안에서도 중위 순회 결과는 정렬된다.

결국 전체 트리에 대해 중위 순회를 하면  
**작은 값들 → 루트 → 큰 값들** 순서가 되어 항상 오름차순으로 방문하게 된다.

**따라서 모든 이진 검색 트리에서 중위 순회 결과는 항상 정렬된 순서이다.**

## 09

높이를 **레벨 수(루트 포함)** 로 보면 다음과 같다.

### 최대 높이
트리가 한쪽으로만 치우친 경우이다.

예:
```text
1
 \
  2
   \
    3
     \
      4
```

이 경우 노드가 n개이면 높이는 `n`이다.

- **최대 높이 = n**

### 최소 높이
가능한 한 균형 있게 채워진 경우이다.  
즉, 높이 `h`인 이진 트리가 가질 수 있는 최대 노드 수는

- `2^h - 1`

이므로, n개의 노드를 담기 위한 최소 높이 `h`는

- `n <= 2^h - 1`
- `n + 1 <= 2^h`
- `h >= log2(n + 1)`

따라서 최소 높이는

- **최소 높이 = `⌈log2(n + 1)⌉`**

---

**답:**
- 최대 높이: `n`
- 최소 높이: `⌈log2(n + 1)⌉`
