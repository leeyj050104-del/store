#3. Heap

```python
import pandas as pd
import heapq

df = pd.read_csv("birthday.csv")

heap = []

for i in range(len(df)):
    try:
        name = str(df.loc[i, "name"]).strip()
        year = int(df.loc[i, "year"])
        month = int(df.loc[i, "month"])
        day = int(df.loc[i, "day"])

        # 가장 늦게 태어난 사람이 먼저 나오도록 음수로 저장
        heapq.heappush(heap, (-year, -month, -day, name))

    except:
        print(f"{name} : 생년월일 정보가 없어 건너뜀")

print("생일이 느린 순서 10명")
for _ in range(10):
    if heap:
        year, month, day, name = heapq.heappop(heap)
        print(f"{name} : {-year}년 {-month}월 {-day}일")
```

#4. 같은 조의 친구들의 생일만 출력하는 코드

```python
import pandas as pd

class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None


class CircularDoublyLinkedList:
    def __init__(self):
        self.head = None
        self.size_val = 0

    def append(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            new_node.next = new_node
            new_node.prev = new_node
        else:
            tail = self.head.prev
            tail.next = new_node
            new_node.prev = tail
            new_node.next = self.head
            self.head.prev = new_node

        self.size_val += 1

    def get(self, index):
        if index < 0 or index >= self.size_val:
            raise IndexError("Index out of range")

        current = self.head
        for _ in range(index):
            current = current.next

        return current.data

    def size(self):
        return self.size_val


df = pd.read_csv("birthday.csv")

my_group = [
    "이시현", "이예지", "이유민", "이윤서", "이주은",
    "이채영", "장채은", "정희원", "조은서", "진가연", "최윤지"
]

cdll = CircularDoublyLinkedList()

for i in range(len(df)):
    name = str(df.loc[i, "name"]).strip()
    year = df.loc[i, "year"]
    month = df.loc[i, "month"]
    day = df.loc[i, "day"]
    cdll.append((name, year, month, day))

print("같은 조 친구들의 이름과 생년월일")

for i in range(cdll.size()):
    name, year, month, day = cdll.get(i)

    if name in my_group:
        try:
            if pd.isna(year) or pd.isna(month) or pd.isna(day):
                raise ValueError("생년월일 없음")
            print(f"{name}: {int(year)}년 {int(month)}월 {int(day)}일")
        except ValueError:
            print(f"{name}: 생년월일 정보가 없습니다.")
```

# 5. Priority Queue Exercises

## 01
**답:** 가능하다.  
최대 힙은 부모 노드의 값이 자식 노드보다 크거나 같다는 조건만 만족하면 된다.  
따라서 서로 조상-자손 관계가 아닌 노드들 사이에서는 값의 크기 관계가 정해지지 않는다.  
즉, 더 얕은 노드가 더 깊은 노드보다 작은 값을 가질 수 있다.

---

## 02
**답:** 아니다.  
최대 힙은 전체 정렬 구조가 아니라 부분 순서 구조이다.  
루트(A[0])가 가장 큰 값인 것은 보장되지만, 마지막 원소가 가장 작은 값이라는 보장은 없다.  
다른 위치에 더 작은 값이 존재할 수 있다.

---

## 03
**답:** ⌈n/2⌉개이다.  
완전 이진 트리에서 리프 노드의 개수는 ⌈n/2⌉개이며, buildHeap()에서는 리프 노드에 대해 스며내리기를 수행하지 않는다.  
따라서 해당 개수만큼의 노드가 검사 없이 넘어간다.

---

## 04
**답:**  
- 최악: Θ(log n)  
- 최선: Θ(1)

최악의 경우 노드가 리프까지 내려가야 하므로 트리 높이만큼 시간이 걸린다.  
반면 이미 힙 조건을 만족하면 한 번도 이동하지 않고 끝나므로 Θ(1)이다.

---

## 05
**답:** 간단하다.  
맨 마지막 원소는 완전 이진 트리 구조를 유지하면서 바로 제거할 수 있다.  
힙의 성질도 깨지지 않기 때문에 별도의 재정렬 과정이 필요 없으며 Θ(1)에 수행된다.

---

## 06
**답:** 아니다.  
이 방법도 힙을 만들 수는 있지만 더 비효율적이다.  

- buildHeap(): O(n)  
- 스며오르기 반복: O(n log n)

따라서 기존 방식이 더 효율적이다.

---

## 07
**답:** 스며오르기(percolateUp)를 사용한다.

과정:
1. 증가한 원소를 부모와 비교한다.
2. 부모보다 크면 교환한다.
3. 교환 후 같은 과정을 반복한다.
4. 루트에 도달하거나 조건을 만족하면 종료한다.

값이 증가하면 부모와의 관계만 깨질 수 있으므로 위쪽으로 이동시키면 된다.  
트리 높이가 log n이므로 시간 복잡도는 O(log n)이다.
